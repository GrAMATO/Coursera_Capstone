## Data

During this project, we will use the [Michelin Restaurants](https://www.kaggle.com/jackywang529/michelin-restaurants) dataset from [Kaggle](https://www.kaggle.com/) and the foursquare API. There is three files, one for each group (one star, two stars and three stars), they all have the same columns names. Here is a little description about theres variables:

- **name**: the restaurant name
- **year**: the year of obtaining the last star
- **latitude**: the restaurant position latitude
- **longitude**: the restaurant position longitude
- **city**: the restaurant's city
- **region**: the restaurant's region
- **zipCode**: zip code of the city
- **cuisine**: type of cooking
- **price**: price of the restaurant, from one \\$ (cheap restaurant) to five \\$ maximum
- **url**: url to the restaurant webpage on the guide michelin website
